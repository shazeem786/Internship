## 1. Scrape the details of most viewed videos on YouTube from Wikipedia:
### Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/
#### You need to find following details:
#### A) Rank
#### B) Name
#### C) Artist
#### D) Upload date
#### E) Views


```python
Url ='https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
```

```python
open_url(Url)

# get webelements for all 30 rows
top30_rows_wbe=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//following::tbody")
# get webelement for table with top 30 YT videos list
# store all text in a list
for i in top30_rows_wbe:
    temp_text=i.text
    break # exit after first iteration
    
sep_rows=temp_text.split('\n')

rank=[]
rank.clear()
vid_name=[]
vid_name.clear()
artist=[]
artist.clear()
views=[]
views.clear()
date=[]
date.clear()

for i in sep_rows:
    
    rank.append(i.split('.')[0])
    
    vid_name.append(i.split('[')[0].split('.')[-1].strip())
    
    artist.append(re.split('\d+',i.split(']')[1])[0])
    
    views.append(str(re.compile('\d+').findall(i.split(']')[1])[0])+'.'+str(re.compile('\d+').findall(i.split(']')[1])[1]))
    
    date.append(str(re.compile('[a-zA-Z]+\s+\d{2}').findall(i.split('.')[2]))+','+str(re.compile('\d{4}').findall(i.split('.')[2])))
```

```python
top30_vid=pd.DataFrame({})
top30_vid['Rank']=rank
top30_vid['Name']=vid_name
top30_vid['Artist']=artist
top30_vid['Views (in billions)']=views
top30_vid['Release Date']=date
```

## 2. Scrape the details team India’s international fixtures from bcci.tv.
### Url = https://www.bcci.tv/.
#### You need to find following details:
#### A) Match title (I.e. 1st ODI)
#### B) Series
#### C) Place
#### D) Date
#### E) Time
#### Note: - From bcci.tv home page you have reach to the international fixture page through code

```python
open_url('https://www.bcci.tv')

time.sleep(5)

# click on "International" button to expand it
driver.find_element_by_xpath("//li[@data-nav-index='0']").click()

# click on 'Fixtures'
driver.find_element_by_xpath("//li[@data-nav-index='0']//following::a[@href='/international/fixtures']").click()

time.sleep(3)

# store webelements for all series names
series_wbe=driver.find_elements_by_xpath("//div[@class='fixture__format-strip']/span")

fixture_frame_text=[]

for i in series_wbe:
    fixture_frame_text.append(i.text)
    
fixture_name=[]
fixture_name.clear()
fixture_type=[]
fixture_type.clear()

time.sleep(3)

for i in range(len(fixture_frame_text)):
    if i%2!=0:
        fixture_name.append(fixture_frame_text[i])
    else:
        fixture_type.append(fixture_frame_text[i])

# get webelment for match title
title_wbe=driver.find_elements_by_xpath("//p[@class='fixture__additional-info']")

title=[]
title.clear()
loc=[]
loc.clear()

for i in title_wbe:
    title.append(i.text.split('\n')[0])
    loc.append(i.text.split('\n')[1])
    
#get webelement for date and time
try:
    date_wbe=driver.find_elements_by_xpath("//div[@class='fixture__datetime desktop-only']")
except:
    pass

date=[]
date.clear()
time=[]
time.clear()

for i in date_wbe:
#     pdb.set_trace()
    date.append(str(i.text.split('\n')[0])+str('-')+str(i.text.split('\n')[1]+str('th')+str(i.text.split('\n')[2])))
    time.append(i.text.split('\n')[3])
```

```python
ind_fixture=pd.DataFrame({})
ind_fixture['Tournament']=fixture_name
ind_fixture['Fixture']=fixture_type
ind_fixture['Match']=title
ind_fixture['Venue']=loc
ind_fixture['Date']=date
ind_fixture['Time']=time
```

## 7. Scrape the details of Data science recruiters from naukri.com.

You have to find the following details:
- A) Name
- B) Designation
- C) Company
- D) Skills they hire for
- E) Location

Note: - From naukri.com homepage click on the recruiters option and the on the search pane type Data science and
click on search. All this should be done through code

```python
driver = webdriver.Chrome(r'C:\Users\admin\Downloads\chromedriver_win32\chromedriver.exe')
# open url
driver.get('https://www.naukri.com/')
# give time to load webpage
driver.set_page_load_timeout(60)
# store main page window handle
main_window = driver.current_window_handle
# remove 2nd window which will be a ad popup
driver.switch_to.window(driver.window_handles[1])
driver.close()
driver.switch_to.window(main_window)
# close the alert
try:
    a = Alert(driver)
    a.dismiss()
except:
    pass
# click on "Recruiters"
driver.find_element_by_xpath("//ul[@class='midSec menu']//following::a[@title='Search Recruiters']").click()

# switch to new tab
driver.switch_to.window(driver.window_handles[1])
# get webelement for search bar
search_wbe = driver.find_element_by_xpath("//div[@class='inpWrap']//following::input[@placeholder='Skills, Designations, Companies']")
# search keyword
search_wbe.send_keys("Data Scientist",Keys.ENTER)
# get recruiters info webelements
vcard_wbe = driver.find_elements_by_xpath("//div[@class='vcard']")
skills_wbe = driver.find_elements_by_xpath("//div[@class='hireSec highlightable']")
# get text from webelements
rec_names = []
rec_names.clear()
rec_roles = []
rec_roles.clear()
rec_companies = []
rec_companies.clear()
rec_locations = []
rec_locations.clear()
skills = []
skills.clear()
# iterating over recruiters info
for i in vcard_wbe:
#     pdb.set_trace()
    if len(i.text.split('\n')) == 4: # if location info is avaialble
        rec_names.append(i.text.split('\n')[0])
        rec_roles.append(i.text.split('\n')[1])
        rec_companies.append(i.text.split('\n')[2])
        rec_locations.append(i.text.split('\n')[3])
    elif len(i.text.split('\n')) == 3: # if location info is missing
        rec_names.append(i.text.split('\n')[0])
        rec_roles.append(i.text.split('\n')[1])
        rec_companies.append(i.text.split('\n')[2])
        rec_locations.append('---')
# iterating over skills sought by recruiters
for i in skills_wbe:
    skills.append(i.text)
```
